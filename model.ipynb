```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,classification_report,confusion_matrix
```

```python
df=pd.read_csv('Wholesale customers data.csv')
```

```python
df.columns
```

```python
df.head()
```

```python
df.groupby('Channel').mean().round(0)
```

```python
# channel is our target variable
# 1- horeca and, 2- retail
import numpy as np
```

```python
X=df.drop(['Channel','Region','Delicassen'],axis=1)
y=df['Channel']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=20)
```

```python
scaling=ColumnTransformer([
    ('scaler',StandardScaler(),X_train.columns)
])
model=Pipeline([
    ('scale',scaling),
    ('naive_baise',GaussianNB())
])
```

```python
model.fit(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model, X, y, cv=50, scoring="f1")
print("Mean f1:", scores.mean())
```

```python
print('accuracy:',accuracy_score(y_test,y_pred))
print("precision",precision_score(y_test,y_pred))
print('recall',recall_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
import pickle
with open('Project_Customer_Category_Classification.pkl','wb') as file:
    pickle.dump(model,file)
```
